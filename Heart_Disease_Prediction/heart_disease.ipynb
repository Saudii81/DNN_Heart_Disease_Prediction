```python
# preprocessing 
categorical_variables = ['sex', 'cp', 'fbs', 'restecg','exang', 'ca', 'thal']
numerics = ['age', 'trestbps','chol', 'thalach', 'oldpeak', 'slope']
```

```python
# one hot encoding for categorical variables
df = pd.get_dummies(df, columns=categorical_variables)
```

```python
df.head()
```

```python
# data splitting 
test_df = df.sample(frac=0.2, random_state=42)
train_df = df.drop(test_df.index) 
```

```python
# checks the training shape 
train_df.shape
```

```python
# checks the test shape 
test_df.shape
```

```python
# calculates mean and standard deviation 
means = train_df[numerics].mean()
sd = train_df[numerics].std()
```

```python
# gets the mean 
means
```

```python
# standardizing the data with means
train_df[numerics] = (train_df[numerics] - means)/sd
```

```python
# standardizing the test data with means and sd
test_df[numerics] = (test_df[numerics] - means)/sd
```

```python
# checks the train dataframe
train_df.head()
```

```python
# converts the data to nunmpy arrays
train = train_df.to_numpy()
test  = test_df.to_numpy()
```

```python
# seperates the features and dependent variables
train_X = np.delete(train, 6, axis=1)
test_X = np.delete(test, 6, axis=1)
```
